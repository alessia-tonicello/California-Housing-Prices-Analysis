VISUALIZE DATA

```python
# A quick way to get a feel of the type of data we are dealing with is to plot a histogram for each numerical attribute
#with plt.savefig, we store the output ( so the figures) in the computer 

#Here we do an HISTOGRAM to start to visualize the data 

import matplotlib.pyplot as plt
housing.hist(bins=50)
plt.tight_layout()
plt.show()

#we have an histogram for each column and shows how data spread around each column 

#MEDIAN_HOUSE_VALUE we have an outlier, we have a lot of houses tht are about a really high value 

#the min and max value of each graph is the min and max value of all the dataset.  
```

```python
#SCATTER PLOT 
#we take the 2 coordinates (longitute and latitude) and we put them on a map

housing.plot(kind="scatter", x="longitude", y="latitude")
plt.tight_layout()
# plt.savefig( "a_first_visualization_plot") -> to save the figure we need this line, in this case we don't
plt.show()

#this map resembles the shape of California; there are main clusters, palces where there are few houses (bay area) and much more in ocean part
#it helps to visualize data 
#in this case we did kind=scatter ad gave the name of the 2 columns x and y: x="longitude", y="latitude"

#these plots use linear scale
```

```python
#The previous plot looks like California (as expected), but other than that it is hard to see any particular pattern
#Setting the alpha option to 0.1 makes it much easier to visualize the places where there is a high density of data points
#alpha is the translucent value 
#alpha = how much light passes through the graph 

housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1)
plt.tight_layout()
# plt.savefig( "a_better_visualization_plot")
plt.show()

#changing the visualization can give us more info about dataset and the output 
#here we have a better idea of where we have aggregation points 
#probably the big clusters are in LA, San Francisco and Sacramento if we compare this plot to a Cali map
#we know the main cities of Cali withouth knowing them, but just knowing the houses sold 
```

```python
#Now we do a variation: we put other info in the scatter plot: population and house values 
#we still use scatter plot with our x,y but here we put a higher alpha value. 
#Then we have s=housing["population"]/100; label = population (=the population value), figsize is the size of the figure 
#"median_house_value", we do a cmap called get -> we import 2 other columns and we are seeing data in a different way

housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
s=housing["population"]/100, label="population", figsize=(10,7),
c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True)
plt.legend()
plt.tight_layout()
# plt.savefig( "housing_prices_scatterplot")
plt.show()

#The output is like doing a correlation without computing it, beucase here:
#population will be in the circles, bigger circle bigger population, so these are the districts (represented by each line), where we have
#the higher number of population

#Many small circles are districts with small houses; on the main cities instead, such as LA and SF, we have higher buildings, so there is a 
#higher number of people

#Then we have the median house value -> the colour of the circle is the house value
#the higher value of houses yields where small circles are -> the houses with higher vlaues are in lower population districts
#so, not skyscrapers but villas and they are outside the part in which we have skyscrapers. 

#The circle size (up right) is population, we enable the legend 

#This plot tells us that the housing prices are very much related to the location (e.g., close to the ocean) and to the population
#density, as we probably knew already
```

DATA CLEANING 

```python
#Most algorithms cannot work with missing features, so we should create a few functions to take care of them

#We delete null values because if you have a big dataset is better to lose some data than process bad data,
#if we need those data we can make a guess, so use some algorithms to put fake data inside those points. 

sample_incomplete_rows = housing[housing['total_bedrooms'].isnull()]
print(sample_incomplete_rows.info())

#we didn't change og dataset but created another one with some other info
#we filter out the rows with total_bedrooms, we took those rows because we can now see those rows and understand if they are meaningful or not
```

There are several strategies to deal with missing values.
The most popular ones are reviewed in [3]

Today, we will consider three strategies:
- Remove the records (i.e., rows) with missing value
- Remove the attribute (i.e., column) altogether
- Replace missing values with some constant value (• e.g., the mean, the median, zero, …) 
- or you can use AI techniques to gather the missing values

```python
#OPTION 1
housing.dropna(subset=["total_bedrooms"])

#here we drop the rows where we have the NULL values
#dropna -> na refers to NULL values
```

```python
#OPTION 2
housing.drop("total_bedrooms", axis=1)

#we drop the column
```

```python
#OPTION 3
median = housing["total_bedrooms"].median()
housing["total_bedrooms"].fillna(median, inplace=True)

#Chanage null value with median value -> we calculate median of the column and then we fill in the blanck spaces with the median value
```

```python
#In this sense we are altering the dataset.
#This is how to clean the dataset for NULL values, so empty values (in some cases we also need ot clean outliers).
```
